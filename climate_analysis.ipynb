# Reflect Tables into SQLAlchemy ORM

```python
#we want to query a SQLite database, and SQLAlchemy is the best tool to do that
# Python SQL toolkit and Object Relational Mapper ORM
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
#SQLAlchemy Create Engine
engine = create_engine("sqlite:///hawaii.sqlite")
```

```python
# SQLAlchemy Automap Base. reflect an existing database into a new model
Base = automap_base()
# SQLAlchemy Reflect Tables. reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# View Classes Found by Automap. We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
#Masurement = Base.classes.measurement
#Station = Base.classes.statio
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Climate Analysis

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results. 

#Starting from the last data point in the database
# Calculate the date one year from the last date in data set.
#prev_yr = dt.date(2017, 8, 23) - dt.timedelta(365=days)
prev_year = dt.date(2017, 8, 23) - dt.timedelta(days=365)

# Perform a query to retrieve the data and precipitation scores
#results = []
#results = session.query(Measurement.date, Measurement.prcp)
#print(results.all())
results = []
results = session.query(Measurement.date, Measurement.prcp)
print(results.all())
```

```python
#filter out all of the data that is older than a year 
#results = session.query(Measurement.date, Measurement,prcp)
#.filter(Measurement.date >= prev_year).all()
results = session.query(Measurement.date, Measurement.prcp).filter(Measurement.date >= prev_year).all()
print(results)
```

```python
# Save the query results as a Pandas DataFrame and set the index to the date column
#df = pd.DataFrame(resuts, columns=['date','precipitation'])
df = pd.DataFrame(results, columns=['date','precipitation'])
df.set_index(df['date'], inplace=True)
print(df)
```

```python
#First we'll convert the DataFrame to strings, and then we'll set our index to "False
print(df.to_string(index=False))
```

```python
# Sort the dataframe by date
#df = df.sort_index()
#print(df.to_string(index=Fasle))
df = df.sort_index()
print(df.to_string(index=False))
```

```python
# Use Pandas Plotting with Matplotlib to plot the data
#df.plot()
df.plot(figsize=(12, 8))
plt.title("Total Precipitation per Day")
plt.ylabel('Inches',fontsize=16)
plt.xlabel('Date', fontsize=16)
plt.xticks (rotation= 90)
```
